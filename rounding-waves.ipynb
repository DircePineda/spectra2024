```python
# Round values
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (example)
data = {
    'wavelength': [530.1, 532.3, 533.5, 534.7, 535.9],
    'intensity': [100, 120, 90, 80, 110]
}

df = pd.DataFrame(data)

# Round wavelength values to integers
df['rounded_wavelength'] = df['wavelength'].round().astype(int)

# Aggregate intensities for each rounded wavelength
result = df.groupby('rounded_wavelength')['intensity'].sum().reset_index()

# Plot original data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(df['wavelength'], df['intensity'], marker='o')
plt.title('Original Data')
plt.xlabel('Wavelength')
plt.ylabel('Intensity')

# Plot rounded data
plt.subplot(1, 2, 2)
plt.plot(result['rounded_wavelength'], result['intensity'],color='red', marker='o', linestyle='-')
plt.title('Rounded Data')
plt.xlabel('Rounded Wavelength')
plt.ylabel('Aggregated Int')
```

```python
# Round values
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (example)
data = {
    'wavelength': [530.1, 532.3, 533.5, 534.7, 535.9],
    'intensity': [100, 120, 90, 80, 110]
}

df = pd.DataFrame(data)

# Round wavelength values to integers
df['rounded_wavelength'] = df['wavelength'].round().astype(int)

# Aggregate intensities for each rounded wavelength
result = df.groupby('rounded_wavelength')['intensity'].sum().reset_index()

# Plot original data
plt.figure(figsize=(8, 6))
plt.plot(df['wavelength'], df['intensity'], marker='o', label='Original Data')

# Plot rounded data
plt.plot(result['rounded_wavelength'], result['intensity'], color='red', marker='o', linestyle='-', label='Rounded Data')

plt.title('Original and Rounded Data')
plt.xlabel('Wavelength')
plt.ylabel('Intensity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Round values & Interpolate intensity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (example)
data = {
    'wavelength': [530.1, 532.3, 533.5, 534.7, 535.9],
    'intensity': [100, 120, 90, 80, 110]
}

df = pd.DataFrame(data)

# Round wavelength values to integers
df['rounded_wavelength'] = df['wavelength'].round().astype(int)

# Interpolate intensity values for rounded wavelengths
rounded_wavelengths = df['rounded_wavelength'].unique()
intensity_values = np.interp(rounded_wavelengths, df['wavelength'], df['intensity'])

# Create a new DataFrame for rounded data
rounded_df = pd.DataFrame({'rounded_wavelength': rounded_wavelengths, 'intensity': intensity_values})

# Plot original data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(df['wavelength'], df['intensity'], marker='o')
plt.title('Original Data')
plt.xlabel('Wavelength')
plt.ylabel('Intensity')

# Plot interpolated data
plt.subplot(1, 2, 2)
plt.plot(rounded_df['rounded_wavelength'], rounded_df['intensity'], color='red', marker='o', linestyle='-')
plt.title('Rounded Data with Interpolated Intensity')
plt.xlabel('Rounded Wavelength')
plt.ylabel('Intensity')

plt.show()
```
